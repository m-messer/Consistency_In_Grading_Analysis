```python
import pandas as pd
import seaborn as sns
```

# Interview Code Analysis

```python
blue = (148/255, 203/255, 236/255)
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('data/interview_code_counts.csv')
df
```

```python
df['Name'] = df['Name'].str.replace('effected', 'affected')
```

```python
code_plt = sns.barplot(df, y='Name', x='Sources', color=blue)
code_plt.set(ylabel='Name', xlabel='Count')
code_plt.get_figure().savefig('plots/interview_code_counts.pdf', bbox_inches='tight')
```
